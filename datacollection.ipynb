```python
import pandas as pd
import numpy as np
import os
import glob
import tabula
```

```python
dfs = tabula.convert_into_by_batch('menus', output_format='csv', pages='all', lattice=True)
```

```python
path = os.getcwd() + '/menus'
csvs = sorted(glob.glob(path + "/*.csv"))
csvs
```

```python
dfs = []
for csv in csvs:
    dfs.append(pd.read_csv(csv).drop('Unnamed: 1', axis=1).drop(0, axis=0).replace('\r', ' ', regex=True))
```

```python
colnames = ['Date', 'DaySin', 'DayCos', 'Meal', 'Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8']
data = pd.DataFrame(columns=colnames)
data
```

```python
row_arch = {'Date': None, 'DaySin': None, 'DayCos': None, 'Meal': None, 'Item1': None, 'Item2': None, 'Item3': None, 'Item4': None, 'Item5': None, 'Item6': None, 'Item7': None, 'Item8': None}
meals = ['Breakfast', 'Lunch', 'Snacks', 'Dinner']
ends = {'Breakfast': (0,8), 'Lunch': (9,17), 'Snacks': (19,21), 'Dinner': (24,33)}
for df in dfs:
    
    year = df.columns[0].split(',')[1].strip()[:4]
    sdate = df.columns[0].split('(')[1].split('-')[0].strip() + year
    sdate = sdate.replace('th', '').replace('rd', '').replace('st', '').replace('nd', '')
    sdate = sdate.replace('Augu', 'Aug ')
    sdate = pd.to_datetime(sdate)
    dates = pd.date_range(sdate, periods=7)
    print(dates)
    
    df.columns = df.iloc[0]
    df.drop(1, axis=0, inplace=True)
    
    for date in dates:
        
        row_arch['Date'] = date
        row_arch['DaySin'] = np.sin(date.dayofweek * (2 * np.pi / 7))
        row_arch['DayCos'] = np.cos(date.dayofweek * (2 * np.pi / 7))
        start = 0
        
        for meal in meals:
            ctr = 1
            row_arch['Meal'] = meal
            for item in df[date.day_name().upper()][ends[meal][0]:ends[meal][1]]:
                row_arch['Item'+str(ctr)] = item
                ctr += 1
           
            data = pd.concat([data, pd.DataFrame([row_arch])], ignore_index=True)
            
            for i in range (1, 9):
                row_arch['Item'+str(i)] = np.nan 
            start = start+ctr
            
            ctr=1
                
data
```

```python
for col in data.columns:
    data[col+'Num'], _ = data[col].factorize()
```

```python
data.drop(['DateNum', 'DaySinNum', 'DayCosNum'], axis=1, inplace=True)
```

```python
data.fillna(-1, inplace=True)
```

```python
data
```

```python
encodings = {}
for col in data.columns[3:12]:
    coldict = {}
    for item in data[col].unique():
        num = data[data[col]==item][col+'Num'].iloc[0]
        coldict[num] = item
    encodings[col] = coldict
```

```python
encodings
```

```python
encodf = pd.DataFrame.from_dict(encodings, orient='index').stack().to_frame()
encodf
```

```python
encodf.to_csv('encodings.csv')
data.to_csv('data.csv')
```
