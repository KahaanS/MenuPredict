```python
import tabula
import pandas as pd
```

```python
dfs = tabula.convert_into_by_batch('pdfs', output_format='csv', pages='all', lattice=True)
```

```python
df = pd.read_csv('pdfs/Weekly Menu 10th July - 16th July.csv').drop('Unnamed: 1', axis=1).drop(0, axis=0)
```

```python
date = df.columns[0].split('(')[1].split('-')[0].strip()
date = date + ' 2023'
date = date.replace('th', '').replace('rd', '').replace('st', '').replace('nd', '')
date = pd.to_datetime(date, format='%d %B %Y')
dates = pd.date_range(date, periods=7)
dates
```

```python
df.columns = df.iloc[0]
df.drop(1, axis=0, inplace=True)
df
```

```python
colnames = ['Date', 'Day', 'Meal', 'Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8']
data = pd.DataFrame(columns=colnames)
data
```

```python
def clear_items(itemdict):
    for i in range (1, 9):
        itemdict['Item'+str(i)] = None
    
```

```python
row_arch = {'Date': None, 'Day': None, 'Meal': None, 'Item1': None, 'Item2': None, 'Item3': None, 'Item4': None, 'Item5': None, 'Item6': None, 'Item7': None, 'Item8': None}
meals = ['Breakfast', 'Lunch', 'Snacks', 'Dinner']
for date in dates:
    row_arch['Date'] = date
    row_arch['Day'] = date.day_name()
    start = 0
    for meal in meals:
        ctr = 1
        row_arch['Meal'] = meal
        for item in df[date.day_name().upper()][start:]:
            if pd.isna(item):
                break
            row_arch['Item'+str(ctr)] = item
            ctr += 1
           
        data = pd.concat([data, pd.DataFrame([row_arch])], ignore_index=True)
        
        for i in range (1, 9):
            row_arch['Item'+str(i)] = None 
        start = start+ctr
        
        for item in df[date.day_name().upper()][start:]:
            if pd.isna(item):
                start+=1
            else:
                break  
    
data
```
